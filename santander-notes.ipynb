```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

#### Load the train and test datasets into memory. The size of each is roughly 300 MB.
    - We don't need the test file for now. We will split the training data.

```python
train_file = "/home/joseph/Dev/datasets/santander-customer-transaction-prediction/train.csv"
#test_file = "/home/joseph/Dev/datasets/santander-customer-transaction-prediction/test.csv"
train_data = pd.read_csv(train_file, index_col=0)
#X_test = pd.read_csv(test_file, index_col=0)
```

#### Lets take a look at the training data.

```python
train_data.head()
```

#### We need to extract the label vector from the data.

```python
y, X = train_data.iloc[:,0], train_data.iloc[:,1:]
```

#### Okay, now let's see some statistics about the data.

```python
X.describe()
```

#### We might need to scale each column, but first let's take a look at the given labels.

```python
y.value_counts()
```

#### So about 10% of the labels are 1 and about 90% are 0. We will need to be careful with our model evaluation.
#### Let's split the data into training and testing data. (It is automatically shuffled or randomly sampled)

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=73, max_iter=1000, tol=1e-3)
sgd_clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, X_train, y_train, cv=3) 
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, y_train_pred)
```
