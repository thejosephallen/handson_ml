## MNIST Data Loading & Setup

```python
# To support both python 2 and python 3
from __future__ import division, print_function, unicode_literals
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1, cache=True)
mnist.target = mnist.target.astype(np.int8)
#sort_by_target(mnist)
mnist
```

```python
X, y = mnist['data'], mnist['target'] # attain the data and the labels
print (X.shape, y.shape) # check the shape
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

an_example = X[40404]
an_example = an_example.reshape(28,28)

plt.imshow(an_example, cmap=matplotlib.cm.binary, interpolation='nearest')
plt.axis('off')
plt.show() 
```

```python
# let's split and shuffle the data
X_train, X_test, y_train, y_test = X[:60000],X[60000:],y[:60000],y[60000:]
random_indices = np.random.permutation(60000)
X_train, y_train = X_train[random_indices], y_train[random_indices]
```

## Binary Classifier
##### Training

```python
# let's classify 4s and not-4s
y_train_4 = (y_train == 4)
y_test_4 = (y_test == 4)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=22)
sgd_clf.fit(X_train, y_train_4)
```

```python
sgd_clf.predict(an_example.reshape(1,-1))
```

#### Evaluation

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, X_train, y_train_4, cv=4, scoring='accuracy')
```

```python
# custom class to predict 4s every time
from sklearn.base import BaseEstimator
class Not4plz(BaseEstimator):
    def fit(self, X, y=None):pass
    def predict(self, X): return np.zeros((len(X),1),dtype=bool)
# how accurate is this?
not4_clf = Not4plz()
cross_val_score(not4_clf,X_train,y_train_4,cv=4,scoring="accuracy")
```

```python
# the dataset is skewed so we need better evalutaion methods (F1)
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, X_train,y_train,cv=4)
y_train_pred.size
# cross_val_predict gives a prediction 
```
