```python
import os, pickle
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})
```

```python
with open(os.getcwd()+"/num_iterations_list.pickle.dat", "rb") as input_file:
    num_iterations_list = pickle.load(input_file)
```

```python
# TEST HOW MANY SAMPLES NEEDED TO SAMPLE EVERY ITEM
# for population sizes from 2 to 200, run 1000 tests to find 
# the avg num of samples of size 2 needed to sample all items
import numpy as np
from random import randint, sample
num_tests = 1000
avg_iterations_list = []
for num_objs in range(2,200):
    x = [i for i in range(num_objs)]
    avg_iterations = 0
    for _ in range(num_tests):
        sampled_count = [0 for _ in range(num_objs)]
        num_iterations = 0
        while 0 in sampled_count:
            i1, i2 = sample(x,2)        
            sampled_count[i1] += 1
            sampled_count[i2] += 1            
            num_iterations += 1
        print num_iterations
        num_iterations_list.append(num_iterations)
        avg_iterations += num_iterations
    avg_iterations /= num_tests
    print "Avg iterations to sample all: ", avg_iterations
    avg_iterations_list.append(avg_iterations)
    
```

```python
plt.plot([i for i in range(2,200)], avg_iterations_list, label = "Avg samples needed vs population size")

plt.xlabel='# Samples Needed'
plt.ylabel='Population Size'
plt.title='Average Samples Needed to Sample Varying Population Sizes'
#plt.grid()

#show the legend
plt.legend()
plt.show()
```
